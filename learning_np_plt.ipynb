```python
sigma = 10
mean = 2
mu = np.ones(1001)*mean

s = np.random.normal(mean,sigma,1000)

count, bins , ignored  = plt.hist(s,1000,density=True)

f = 1/np.sqrt(2*(np.pi)*(sigma**2))*np.exp(-((bins-mu)**2)/(2*(sigma**2)))

plt.plot(bins,f,linewidth=4,color='r')
```
