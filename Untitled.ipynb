## Determining the fatigue crack propogation curves 

$$
\frac{da}{dN} = C\Delta K^m
$$

```python
delta_K = np.linspace(300,1200,100)
m01 = 5.0585
m05 = 5.5347
m = 4.6899

C01 = 1.5624e-19
C05 = 3.0907e-20
C = 2.4329e-18

da_dN01 = C01*(delta_K**m01)
da_dN05 = C05*(delta_K**m05)
da_dN = C*(delta_K**m)
plt.subplot(2,1,1)
plt.yscale('log')
plt.xscale('log')
plt.plot(delta_K,da_dN01,'y--',label="R=0.1")
plt.plot(delta_K,da_dN05,'g--',label="R=0.5")
plt.plot(delta_K,da_dN,label="R=0.1+R=0.5")
plt.legend()

# fake data
r = np.random.normal(1,0.1,size=100)
plt.subplot(2,1,2)
plt.yscale('log')
plt.xscale('log')
plt.plot(delta_K*r,da_dN01,'y^',label="R=0.1")
plt.plot(delta_K*r,da_dN05,'ro',label="R=0.5")
plt.plot(delta_K,da_dN,label="R=0.1+R=0.5")
plt.legend()
```
